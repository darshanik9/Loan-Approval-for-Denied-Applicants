```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
raw_data_train = pd.read_csv("train_loan.csv")
raw_data_test = pd.read_csv("test_loan.csv")
```

```python
print(raw_data_train.describe(include = 'all'))
print('\n SHAPE OF TRAIN DATA SET:',raw_data_train.shape)
```

```python
print(raw_data_test.describe(include = 'all'))
print('\n SHAPE OF TEST DATA SET: ',raw_data_test.shape)
```

## Combining Train and Test Data sets for treating Missing values

```python
na = raw_data_train.shape[0]
nb = raw_data_test.shape[0]
comb_data =  pd.concat([raw_data_train, raw_data_test], sort = False).reset_index(drop = True)
#Setting Dependent variable to y_train
y_train = comb_data['Loan_Status'].to_frame()
#Dropping ID and dependent variable from combined dataset
comb_data.drop(['Loan_Status','Loan_ID'], axis = 1, inplace = True)
print('\n Shape of Combined Dataset after merging train and test:', comb_data.shape)
print('\n Shape of Train Dataset after merging train and test:', na)
print('\n Shape of Test Dataset after merging train and test:', nb)
```

## Verifying Missing values and Percentage of Missing values in each variable 

```python
Percent_missing = (comb_data.isnull().sum()/comb_data.isnull().count())*100
print('\n Percent of Missing values')
print('\n', Percent_missing)
print('\n Missing values')
print('\n', comb_data.isnull().sum())
```

```python
y_train['Loan_Status'] = y_train['Loan_Status'].map({'Y':1, 'N':0})
cols = ['LoanAmount','ApplicantIncome','CoapplicantIncome','Credit_History']
sns.set(color_codes = True)
sns.pairplot(comb_data[cols], kind = 'reg', height = 2.0, aspect = 1.5,)
```

```python
def dist(variable_1,variable_2,variable_3):
    sns.displot(comb_data[variable_1])
    sns.displot(comb_data[variable_2])
    sns.displot(comb_data[variable_3])
dist('LoanAmount','ApplicantIncome','CoapplicantIncome')
```

```python
comb_data['Credit_History']
```

```python
COL1 = ['ApplicantIncome','CoapplicantIncome','LoanAmount']
for a in [COL1]:
    comb_data[a] = comb_data[a].fillna(comb_data[a].median())
```

```python
comb_data['Credit_History'].mode()
```

```python
COL2 = ['Credit_History']
for b in [COL2]:
    comb_data[b] = comb_data[b].fillna(1.0)
```

```python
COL3 = ['Loan_Amount_Term']
for c in [COL3]:
    comb_data[c] = comb_data[c].fillna(360)
```

```python
y_train = y_train.map({'Y':1, 'N':0})
```

```python
comb_data_enc = pd.get_dummies(comb_data)
print('\n Shape of combined data set before encoding:',comb_data.shape)
print('\n Shape of combined data set after encoding:',comb_data_enc.shape)
print('\n No. of Newly Encoded Variables:', comb_data_enc.shape[1]-comb_data.shape[1])
```

```python
train_final = comb_data_enc[:na]
test_final = comb_data_enc[na:]
```

```python
def outliers():
    fig = plt.figure(figsize = (20,20))
    ax1 = plt.subplot2grid((5,3),(0,0))
    plt.scatter(x = train_final['ApplicantIncome'], y = train_final['Loan_Status'], color = 'skyblue')
    plt.axvline(x = 80000, color = 'red')
    plt.title('Applicant Income and Loan_Status Scatter Plot')

    ax1 = plt.subplot2grid((5,3), (0,1))
    plt.scatter(x = train_final['CoapplicantIncome'], y = train_final['Loan_Status'], color = 'violet')
    plt.axvline(x = 30000, color = 'red')
    plt.title('Co-Applicant Income and Loan Status Scatter Plot')

    ax1 = plt.subplot2grid((5,3), (1,0))
    plt.scatter(x = train_final['LoanAmount'], y = train_final['Loan_Status'], color = 'green')
    plt.title('Loan Amount and Loan Status Scatter Plot')

    ax1 = plt.subplot2grid((5,3), (1,1))
    plt.scatter(x = train_final['Loan_Amount_Term'], y = train_final['Loan_Status'], color = 'orange')
    plt.title('Loan term and Loan Status Scatter Plot')

    ax1 = plt.subplot2grid((5,3), (2,0))
    plt.scatter(x = train_final['Credit_History'], y = train_final['Loan_Status'], color = 'blue')
    plt.title('Credit History - Loan Status Scatter Plot')
```
